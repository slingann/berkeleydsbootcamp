### Question 4 (Absolute Addition)
Implement the function abs_addition that computes the absolute value of a + the absolute value of b.
Hint: If a is negative, the absolute value of a is -a. Similarly if b is negative, the absolute value of b is -b.

**abs_addition(-2,-3)**
>5
**abs_addition(1, -5)**
>6

```python
def abs_addition(a, b):
    if a < 0:
        abs_a = -a
    else:
        abs_a = a
    if b < 0:
        abs_b = -b
    else:
        abs_b = b
    return abs_a + abs_b
```

```python
print(abs_addition(-2, -3))#should print 5
print(abs_addition(1, -5))#should print 6
```
